## Intro. to Probability

```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn2
```

### 1) Probability

$$\huge P(A) = \frac{\text{no. of outcomes favorable for event A}}{\text{total number of possible outcomes}}$$

- *Where $P(A) = 0$ is when the event doesn't occurr and if $P(A) = 1$ we certainly know that the event is taking place*
- The number of possible outcomes, for example, when tossing one coin for event $A$, the sample space is $\{{\frac{1}{2}, \frac{1}{2}} \}$

What is the probability of getting tails when tossing a coin?

$\huge \begin{array}{c}{{P(H)=\frac{1}{2}}}\\ {{P(T)=\frac{1}{2}}}\end{array}$

```python
venn2( (1/2, 1/2, 0), set_labels = ('Head', 'Tail') );
```

- Both events have an equal chance of happening, but **however both events can never happen at the same time**, so they are mutually exclusive or disjoint events

### 2) Complementary of Probability

If we have the case of the coin, and the probability that an event will certainly occur is 1, then if we subtract 1 minus the probability that one of the possible events will occur, it gives the probability of the other remaining events taking place.

$\huge{P(\bar{H})} =\huge P(\text{Not H}) = 1 - P(T) = 1 - 0.5 = 0.5$

Again, we see that since there are only 2 possible events, the complementary of one is the probability of the other remaining happening.

### 3) Union of two events (logical OR)

Probability of one event or the other occurring

$\huge P(A \cup B)$

#### 3.1) For disjoint events

```python
venn2( (1/2, 1/2, 0), set_labels = ('Head', 'Tail') );
```

**These kind of events never take place at the same time**

When the events are mutually exclusive, that is, for example, the probability that when tossing a coin comes up heads or tails is 100% since we will always obtain one of the two outcomes.

$\huge P(A \cup B) = P(A) + P(B)$

Where $A$ and $B$ are events that doesn't have any outcome in common

#### 3.2) For joint events

```python
venn2( (0.80, 0.70, 0.20) );
```

**These kind of events can take place at the same time**

An example of this would be the probability of rain or wind, both can happen at the same time.

Imagine that the probability of raining is $P(A) = 0.8$ and wind, $P(B) = 0.7$, **if we just do the sum to get the union like disjoint events, we get a probability number greater than 1, which doesn't exists**

- In fact, to calculate the probability of raining OR wind, we need to substract the probability of both occurring (intersection)

$\huge P(A \cup B) = P(A) + P(B) - P(A \cap B)$

### 4) Intersection of two independent events (logical AND)

Two **events are independent when the outcome of event A does not affect or condition the outcome of event B** or vice versa.

```python
venn2( (0.5, 0.5, 0.25) );
```

For example, if we want to know the probability of obtaining the secuence Head, Head (two consecutive heads) when tossing a coin, both 2 events, first toss, and second toss, are independent since getting head in the first try doen't mean you'll also get heads at the next time

$\huge P(A \cap B) = P(A) \cdot P(B)$

*(product rule) for events $A$ and $B$ which are not related between each other*

#### Exercise

Imagine you are with 50 friends at a party, given the 365 days of a year, **which is the probability that everyone has a different birthday?**

- Given the sample space: $S = \{ 1,2,...364,365 \}$, we know that we are looking for the intersection of each birthday event

- Since one friend's birthday doen't affect the birthday of his other friend, these events are **independent**

- Using the product rule: $\large P(A \cap B) = P(A) \cdot P(B) = \prod_{i=1}^{364} (\frac{365 - i}{365}) = \frac{364}{365} \cdot \frac{363}{365} \cdot \frac{362}{365} \cdot \frac{361}{365}...
$

- This is because the probability $X_i$ for a friend of having a birthday in a different day to the rest is lower as we increase the number of friends in the party, as the numerator decreases, "there are less day avaiable to randomly pick"

```python
def birthday_prob(n_friends):

    total_prob = 1

    for i in reversed(range(365-n_friends, 366)):
        #starts i=364 and keeps on depending on number of events (no. of people)
        total_prob*=(i)/(365)
    
    return total_prob
```

```python
birthday_prob(n_friends=9)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_friends = np.arange(2, 300)
```

```python
probabilities = [birthday_prob(n) for n in n_friends] 
```

```python
plt.xlabel("No. of friends in the party")
plt.title("P(not overlapping birthdays)")
plt.axhline(y=0.5,color="orange")
plt.scatter(n_friends, probabilities);
plt.show()
```

If there are around 23 people at the party, it will be more likely that they will have a birthday in common.
